```python
# #xample

class Avenger:

    def __init__(self, name, weapon, nationality="American"):
        self.name = name
        self.weapon = weapon
        self.nationality = nationality


    def introduction(self):
        print(f" Hello my name is {self.name}")

    def carrying_weapon(self):
        print(f" My weapon is {self.weapon}")

# Creating an Object

Iron_man = Avenger("Iron Man", "Suit")
print(Iron_man.introduction())
print(Iron_man.carrying_weapon())
```

### Creating a Class for Cartoon Charaters

```python
class Cartoon:
    def __init__(self, name, show, catchphrase):
        self.name = name
        self.show = show 
        self.catchphrase = catchphrase

    def introduction(self):
        print(f"Hello My name is {self.name}")

    def speak(self):
        print(f"{self.name} says {self.catchphrase}")
```

```python
Mickey_Mouse = Cartoon("Mickey Mouse", "Mickey Mouse Clubhouse", "Oh Boy")
print(Mickey_Mouse.show)
print(Mickey_Mouse.catchphrase)
print(Mickey_Mouse.name)
```

```python
Spongebob = Cartoon("SpongeBob Squarepants", "Spongebob", "I'm Ready")
print(Spongebob.name)
print(Spongebob.show)
print(Spongebob.introduction())
print(Spongebob.speak())
```
