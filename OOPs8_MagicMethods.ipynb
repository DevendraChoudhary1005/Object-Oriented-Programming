## Magic Methods

#### Magic methods are also called as Dunder Methods because they have double underscores_before and after and they are special methods in python that allows us to customise how objects behave.

#### These methods can help us define:
#### 1). Object Initialization (__init__)
#### 2). String Representation (__str__, __repr__)
#### 3). Arithimetic Operations (__add__, __sub__, etc)
#### 4). Comparison Operations (eq, lt)
#### 5). Length, Indexing and iteration (__len__, __getitem__, etc)

#### Why do we use Magic Methods
#### 1. Makes objects behave like built in types (eg, lists, integers)
#### 2. Improves Readability by allowing basic operations object1 + object2
#### 3. Simplifies code by replacing verbose function calls with natural syntax.

### Operator Overloading
#### It allows us to define how the operators (+,-,*,/,etc) work with user defined objects.
#### Why is operator overloading Useful?
#### makees custom objects behave like built in types (numbers, strings, lists, etc)
#### It improves readability by using natural syntax (object1 + object2) instead of object1.add(object2)

#### Operator --> __+__ --> Method Name --> __add__(self,other)
#### Operator --> __-__ --> Method Name --> __sub__(self,other)
#### Operator --> __*__ --> Method Name --> __mul__(self,other)
#### Operator --> __/__ --> Method Name --> __truediv__(self,other)
#### Operator --> __==__ --> Method Name --> __eq__(self,other)
#### Operator --> __>__ --> Method Name --> __gt__(self,other)

```python
class Avenger:
    def __init__(self, name, power):
        self.name = name
        self.power = power

    def __add__(self, other):
        return Avenger(self.name + " & " + other.name, self.power + other.power)

    def __sub__(self, other):
        return Avenger(self.name, self.power - other.power)

    def __mul__(self, other):
        return Avenger(self.name, self.power * other.power)

    def __truediv__(self, other):
        return Avenger(self.name, self.power / other.power)

    def __str__(self):
        return (f"{self.name} has {self.power} power")
        
```

```python
thor = Avenger("Thor", 200)
hulk = Avenger("Hulk", 190)
ironman = Avenger("Iron Man", 150)
```

```python
team = thor + hulk
print(team)

team = thor - hulk
print(team)

team = thor * hulk
print(team)

team = thor / hulk
print(team)
```

```python
# Example 2

class AIModel:
    def __init__(self, name, accurracy):
        self.name = name
        self.accurracy = accurracy

    def __gt__(self, other):
        return self.accurracy > other.accurracy
    
    def __lt__(self, other):
        return self.accurracy < other.accurracy

    def __eq__(self, other):
        return self.accurracy == other.accurracy
```

```python
claude = AIModel("Claude", 0.9)
chatgpt = AIModel("ChatGPT", 0.8)
davinci = AIModel("Davinci", 0.7)
```

```python
print(claude > chatgpt)
```

```python
print(claude < davinci)
```
