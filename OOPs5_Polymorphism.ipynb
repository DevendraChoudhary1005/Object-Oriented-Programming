## Polymorphism
#### Polymorphism means "Many Forms" in Greeks. So, here it allows objects of different classes as they are the objects of one class and it enables reusability and makes program more flexible.
#### It allows different data types to be handled using a single, uniform interface.
#### Think of it like a universal remote control. The "Power" button is a single interface, but it performs a different action depending on whether you are pointing it at a TV, a fan, or an air conditioner.

### Types of Polymorphism:
#### 1). Method Overriding (Runtime Polymorphism)
#### 2). Method Overloadinng (Compile time Polymorphism)
#### 3). Operator Overriding

### 1). Method Overriding (Runtime Polymorphism): 
#### The specific method to be executed is determined at runtime based on the actual object being referenced. This happens when a child class provides a specific implementation for a method that is already defined in its parent class.
#### How it works: You use a reference variable of the parent class to point to an object of the child class.

#### The Rule: The method in the child class must have the same name, return type, and parameters as the one in the parent class.

#### This allows dynamic selection at runtime.

```python
# Example of Runtime Polymorphism

class GrandParent:      #Parent Class

    def speak(self):
        print("Hello My name is DC")

class Parent(GrandParent):      # Child Class

    def speak(self):
        print("Hello my name is DC Junior")

class Child(GrandParent):

    def speak(self):
        print("Hello my name is Choudhary")
```

```python
def attribute_provider(objects):
    return objects.speak()
```

```python
grandpa = GrandParent()
```

```python
print(attribute_provider(grandpa))
```

```python
parent = Parent()
```

```python
print(attribute_provider(parent))
```

```python
# Example 2

class Shape:

    def area(self):
        print("This method return the area of the shape.")

class Rectangle(Shape):

    def __init__(self, height, width):
        self.height = height
        self.width = width
    
    def area(self):
        return (self.width*self.height)
    
class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14*self.radius*self.radius
```

```python
shape = Shape()
```

```python
shape.area()
```

```python
rectangle = Rectangle(10,20)
```

```python
rectangle.area()
```

```python
circle = Circle(20)
```

```python
circle.area()
```
