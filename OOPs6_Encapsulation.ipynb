## Encapsulation

#### It allows different data types to be handled using a single, uniform interface.
#### Encapsulation is a way of wrapping data (variables and Methods) inside a capsule or a class as a single unit.
#### It restricts the direct access to some of the object component's which is a mean of perventing accidental inference and misuse of the data.

### Access Modifiers (Public, Private and Protected):
#### In Object-Oriented Programming, Access Modifiers (Public, Private, and Protected) act like security clearances. They determine who can see or change the data inside a class.

### Key Concepts of Encapsulation:
#### 1). Public Members
#### 2). Protected Members
#### 3). Private Members
#### 4). Gatter and Setter Methods

#### 1. Public Member: Accessible from anywhere or attributes are accessible outside of the class.
#### And these member are accessible by the derived class.
#### This doesn't need any syntax to be declared, any method is public by default. 

```python
# Example

class Ironman:
    def __init__(self, name, power):
        self.name = name
        self.power = power
        
    def introduction(self):
        print(f"Hello My name is {self.name} and my power is {self.power}")
        
```

#### dir() --> it takes the input as the object and gives us this idea that which member is public or private or protected.

```python
Ironman_1 = Ironman("Ironman", "Suit")
Ironman_1.introduction()
```

```python
dir(Ironman_1)
```

#### 2. Private Member : A private member is not accessible outside the class and not even in the derived class and in order to have the access over these private members or to bring some modification on these we use the gatter and shatter methods.
#### Private members are declared by using the double underscore "__" before the name.


```python
class Avengers:
    def __init__(self, name, phone_number):
        self.name = name
        self.__phone_number = phone_number   # Phone number is a private member
```

```python
avenge = Avengers("Ironman", 123456789)
print(avenge.name)
# print(avenge.__phone_number)  # This will raise an AttributeError because phone_number is a private member
# To show the private member
print(avenge._Avengers__phone_number)
```

#### Python doesn't actually make the variable "invisible." Instead, it performs Name Mangling. It renames the variable behind the scenes using this specific formula:
#### _ClassName__variable_name

```python
dir(avenge)  
```

#### In Encapsulation, we make our variables private (using __) so that they cannot be changed directly from outside the class. However, we still need a way to see or update that data safely. That’s where these methods come in.

### Getter Function (The Read Access)

#### It will be helping us out with the accessing the private data.
#### It lets you control how the data is presented (e.g., formatting a phone number or checking if the user has permission to see it).

```python
class Captain:
    def __init__(self, name, address):
        self.name = name
        self.__address = address

    # Getter Method
    def access_address(self):
        print(f"My Address is {self.__address}")
```

```python
captain = Captain("Captain America", "Brooklyn")
```

```python
# captain.address    # this won't work
captain.access_address()
```

### Setter Function (The "Write" Access)
#### A setter method allows you to change the value of a private variable.
#### This is the most important part. It allows you to validate the data before saving it. You can stop someone from setting a "price" to a negative number or an "age" to 500.

```python
class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.__sound = sound

    # Getter Method
    def get_sound(self):
        print(f"The {self.name} says {self.__sound}")

    # Setter Method
    def set_sound(self, new_sound):
        self.__sound = new_sound  # Update the sound
        print(f"The New Sound of {self.name} is {self.__sound}")
```

```python
animal = Animal("Dog", "Woof-Woof")
```

```python
animal.name
```

```python
animal.get_sound()
```

```python
# Now to update the sound
animal.set_sound("Bark-Bark")
```

#### 3. Protected Member: It is the data which is not accessible outside the class but the protected data is going to be accessible by derived class or the child class.
#### For having the protected member you can mention your protected member as "_attribute_name" or by using single underscore in front.

```python
class GrandParent:
    def __init__(self, name, locker_key):
        self.name = name
        self._locker_key = locker_key

    # Getter Method
    def locker_info(self):
        print(f"The Locker Key of {self.name} is {self._locker_key}")
        
```

```python
# Child Class

class Parent(GrandParent):
    def __init__(self, name, locker_key):
        super().__init__(name, locker_key)

    def child_reveals(self):
        print(f"The Locker Key information is {self._locker_key}")
```

```python
grandpa = GrandParent("Grandpa", 1234)
```

```python
grandpa.name
```

```python
grandpa.locker_info()
```

```python
parent = Parent("Parent", 5678)
```

```python
# parent.locker_key won't work
parent.child_reveals()
```

#### Example for all Access Modifiers

```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.__balance = balance

    # Getter Method
    def get_balance(self):
        print(f"The Total Balance of {self.account_holder} is {self.__balance}")

    # setter method
    def deposits(self, amount):
        if amount>0:
            self.__balance += amount
            print(f"The Deposited amount is {self.__balance}")
            print(f"The Total Balance of {self.account_holder} is {self.__balance}")
        else:   
            print("Invalid Amount")

    def withdraw(self, withdrawal_amount):
        if withdrawal_amount>0 and withdrawal_amount <= self.__balance:
            self.__balance -= withdrawal_amount
            print(f"The Withdrawal amount is {withdrawal_amount}")
            print(f"The Total Balance of {self.account_holder} is {self.__balance}")
        elif withdrawal_amount>self.__balance:
            print("Insufficient Balance")
        else:
            print("Invalid Amount")
```

```python
account = BankAccount("Devendra", 85000000000)
```

```python
account.account_holder
```

```python
# account.__balance
account.get_balance()
```

```python
account.deposits(85000000)
```

```python
account.withdraw(95000000)
```
