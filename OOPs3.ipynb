## Pillars of OOPs

### Inheritance:
#### In Object-Oriented Programming, Inheritance is a mechanism that allows one class (the Child or Subclass) to derive its attributes and methods from another class (the Parent or Superclass).
#### It is a fundamental conept in OOP because of which the child class is able to inherit the properties OR attributes or functioalities of the Parent class.
#### Example - Just like a child is going to inherit the property from parents and parents inherited the property form the grandparents. And definitly  in this hierarchy the grandchild can have some features different from the parent and grandparents as well.

#### In Python Typically we get to see 3 types of Inheritance:
#### 1) Single Inheritance - This is the simplest form. A single child class inherits from a single parent class.
#### 2) Multilevel Inheritance - A child class acts as a parent for another child class. It creates a "grandparent-parent-child" chain.
#### 3) Multiple Inheritance - A single child class inherits from more than one parent class. This allows the child to combine features from different sources.

```python
"""
1) Single Inheritance
Parent Class (Properties --> Attributes and functionalities)
    |
    |
Child Class
"""

class Avenger:
    def __init__(self, name, power):
        self.name = name
        self.power = power

    def introduction(self):
        print(f"My name is {self.name}")

class IronMan(Avenger):
    def __init__(self, name, power, language):     # We have to mention the functions of parent class (name, power)
        super().__init__(name, power) #The super() function is a built-in Python tool used to give a child class access to the methods and properties of its parent class
        self.language = language

    def language_detail(self):
        print(f"My native language is {self.language}")
```

#### When you create a child class, it often needs to do everything the parent does plus a few extra things. If you define a new __init__ in the child, it overwrites the parent's constructor. super() allows you to run the parent's setup code before adding your own.

```python
"""
2) Multilevel Inheritance:
Grand Parent Class (Properties)
    |
Parent Class (Properties)
    |
Child Class
"""

class GrandParent:
    def __init__(self, nationality, baldness, anger_issues):
        self.nationality = nationality
        self.baldness = baldness
        self.anger_issues = anger_issues

    def introduction(self):
        print(f"Hello my nationality is {self.nationality} and about my anger issues you want to ask about then my answer is {self.anger_issues}")
```

```python
class Parent(GrandParent):
    def __init__(self, nationality, baldness, anger_issues, language):
        super().__init__(nationality, baldness, anger_issues)
        self.language = language

    def language_details(self):
        print(f"My Mother togue is {self.language}")
```

```python
class Child(Parent):
    def __init__(self, nationality, baldness, anger_issues, language, job):
        super().__init__(nationality, baldness, anger_issues, language)
        self.job = job

    def job_details(self):
        print(f"I am a {self.job} at a major Tech Company and I make millions per year")
```

```python
# Creating object for Parent class
parent_1 = Parent("Indian", "Yes", "No", "Hindi or Marwadi")
```

```python
parent_1.introduction()
```

```python
# Creating object for CHild Class
Child_1 = Child("Indian", "No", "Yes", "Hindi or English", "Data Scientist")
```

```python
Child_1.introduction()
```

```python
Child_1.job_details()
```

```python
# Creating Object for GrandParent Class
Grand_Parent = GrandParent("Indian", "Yes", "Yes ")
```

```python
"""
3) Multiple Inheritance:
Parent_1 + Parent_2 + Parent_3 ...... 
                |
            Child Class
"""
class IronMan:
    def __init__(self, name, power, suit, weapon, nationality="American"):
        self.name = name
        self.power = power
        self.suit = suit
        self.weapon = weapon
        self.nationality = nationality

    def introduction(self):
        print(f"My name is {self.name} and power is {self.power}")

class Cap_America:
    def __init__(self, name, power, weapon, nationality = "America"):
        self.name = name
        self.power = power
        self.weapon = weapon
        self.nationality = nationality

    def weapon_details(self):
        print(f"My weapon is {self.weapon}")
```

```python
# Now creating a child class which is inherited from both the parent class.
class Hulk(IronMan, Cap_America):
    def __init__(self, name, power, suit, weapon, nationality, anger_issues):
        super().__init__(name, power, suit, weapon, nationality)
        self.anger_issues = anger_issues

    def anger_des(self):
        
        print(f"About my anger issues answer is {self.anger_issues}")
```

```python
# Creating child class Object
hulk = Hulk("Hulk", "Amazing strength", "Mark 85", "Gamma Rays", "American", "Yes")
```

```python
hulk.introduction()
```

```python
hulk.weapon_details()
```

```python
hulk.anger_des()
```
