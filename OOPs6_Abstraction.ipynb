### Abstraction

#### Abstraction is the process of hiding the complex internal details of how something works and only showing the essential features to the user.
#### So, this helps in reducing the programming complexity and efforts.

#### Abstraction allows us to work with an object with what an object does rather than how it does

#### 1) Abstract class - It is an class that cannot be instantiates or that means for an abstract class we can't create the object of the class.
#### 2) Abstract method - Abstract method just because it can't be accessed by the object of the abstract class itself thats why we have to call the abstract method in our child class and the implementation of the abstract method is not provided inside the abstract class but must be provided in the class.

#### The abstract method is being defined by the @abstractmethod


```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car started")

    def stop(self):
        print("Car stopped")

class ElectricScooter(Vehicle):
    def start(self):
        print("Electric scooter started")

    def stop(self):
        print("Electric scooter stopped")
```

```python
car = Car()
```

```python
car.start()
```

```python
car.stop()
```

```python
iqube = ElectricScooter()
```

```python
iqube.start()
```

```python
iqube.stop()
```
