### Different Types of Constructors:
#### 1) Default Constructor - it doesn't have the parameters inside it and it is only having the self keyword inside it. It assigns a default value to all the objects.
#### 2) Parameterised COnstructor - It consists of Parameters inside it so that every object can have a speicfic attribute associated to it. It allows us to pass arguments while creating an object.

```python
# Example of Default Constructor:

class Dog:
    def __init__(self):
        self.name = "Labrador"
        self.color = "Brown"

    def show_info(self):
        print(f"This dog is {self.name} and the color of the dog is {self.color}")
        
```

```python
dog1 = Dog()
```

```python
dog1.name
```

```python
dog1.show_info()
```

```python
dog2 = Dog()
```

```python
dog2.name
```

#### Since, here we can see that a default value is assigned to all the objects. (Because of Default Constructor)

```python
# Example of Parameterised Constructor

class Dogs:
    def __init__(self, breed, color):
        self.breed = breed
        self.color = color

    def show_info(self):
        print(f"The Dog is {self.breed} and is having color as {self.color}")
```

```python
dog1 = Dogs("Golden Retriver", "Golden")
dog1.show_info()
```

```python
dog2 = Dogs("Husky", "White")
dog2.show_info()
```

### Methods --> a method is simply a function that belongs to a class

### Types of Methods
#### 1) Instance Method - They are designed to act on a specific object (an instance). These operate on Object Attributes. The  Instance Method is going to use self keyword to access and modify object attribute.
#### 2) Static Method - These belong to the Class itself, not to any one object. If you change something here, it affects the "blueprint. They use the @classmethod decorator and take "cls" (the class) as the first argument instead of self. And these are called using the class name as well.
#### 3) Class Method - These are "independent" functions that just happen to live inside a class for organization. They don't know anything about the object or the class. They use the @staticmethod decorator. Static method can also be called by class name.

```python
# 1) Instance Method :

class Avengers:
    def __init__(self, name, power):
        self.name = name
        self.power = power
    
    def show_details(self):  # this is a instance method
        print(f"I am {self.name} and my power is {self.power}")

```

```python
iron_man = Avengers("Iron man", "Suit")
iron_man.show_details()
```

```python
cap_America = Avengers("Captain America", "Shield")
cap_America.show_details()
```

```python
# 2) Class Method

class Avengerss:

    # Now we create an attribute outside of the constructor
    origin = "Avenger"   # Now this attribute is common for all the objects of this class

    def __init__(self, name, power):
        self.name = name
        self.power = power

    @classmethod   # This is a class method
    def origin_info(cls):
        print(f"My Origin is {cls.origin}")

    def self_info(self):
        print(f"My name is {self.name} and my power is {self.power}")

```

```python
Thor = Avengerss("Thor", "Hammer")
Thor.self_info()
```

```python
Thor.origin_info()       # this doesn't deppend on the object
```

```python
# 3) Static Method: 

class Avengersss:

    origin = "Avenger"

    def __init__(self, name, power):   
        self.name = name
        self.power = power

    def introduction(self):  # This is a Instance Method and here we use self attribute
        print(f"My name is {self.name} and my power is {self.power}")

    @classmethod # this is a class method caller
    def origin_info(cls):   # here we need cls attribute
        print(f'I am an {cls.origin}')

    @staticmethod     # this is a static method caller
    def mission():     # this method doesn't need self or cls
        print("All Avenger must protect the Earth")
```

```python
Iron_man = Avengersss("Iron man", "Suit")
```

```python
Iron_man.introduction()
```

```python
Iron_man.origin_info()
```

```python
Iron_man.mission() # this static method can be called with Object name 
```

```python
Avengersss.mission()   # also static method can be called using the class name
```
